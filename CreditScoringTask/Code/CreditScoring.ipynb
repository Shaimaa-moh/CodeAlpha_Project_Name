```python
train_data.dropna(axis=1, thresh=len(train_data)*0.7, inplace=True)  # Drop columns with more than 30% null values
train_data.isnull().sum()
```

```python
drop_columns=['ID',"Name","SSN",'Payment_Behaviour',"Type_of_Loan","Age","Month","Occupation"]
train1=train_data.drop(drop_columns,axis=1)
```

```python
# Deleting (-) and (_) in Column
train1['Annual_Income'] = train1['Annual_Income'].str.replace('-','')
train1['Annual_Income'] = train1['Annual_Income'].str.replace('_','')

# Deleting (-) and (_) in Column
train1['Num_of_Loan'] = train1['Num_of_Loan'].str.replace('-','')
train1['Num_of_Loan'] = train1['Num_of_Loan'].str.replace('_','')

# Deleting (-) and (_) in Column
train1['Changed_Credit_Limit'] = train1['Changed_Credit_Limit'].str.replace('-','0')
train1['Changed_Credit_Limit'] = train1['Changed_Credit_Limit'].str.replace('_','0')

# Deleting (-) and (_) in Column
train1['Outstanding_Debt'] = train1['Outstanding_Debt'].str.replace('-','')
train1['Outstanding_Debt'] = train1['Outstanding_Debt'].str.replace('_','')

# Deleting (-) and (_) in Column
train1['Monthly_Balance'] = train1['Monthly_Balance'].str.replace('-','')
train1['Monthly_Balance'] = train1['Monthly_Balance'].str.replace('_','')


# Deleting (-) and (_) in Column
train1['Num_of_Delayed_Payment'] = train1['Num_of_Delayed_Payment'].str.replace('-','')
train1['Num_of_Delayed_Payment'] = train1['Num_of_Delayed_Payment'].str.replace('_','')

train1['Amount_invested_monthly'] = train1['Amount_invested_monthly'].str.replace('-','')
train1['Amount_invested_monthly'] = train1['Amount_invested_monthly'].str.replace('_','')


train1['Credit_Mix'] = train1['Amount_invested_monthly'].str.replace('-','')
train1['Credit_Mix'] = train1['Amount_invested_monthly'].str.replace('_','')

# train1['Amount_invested_monthly'] = train1['Amount_invested_monthly'].str.replace('_','')
# train1['Amount_invested_monthly'] = train1['Amount_invested_monthly'].astype(float)
```

```python

train1['Annual_Income'] = train1['Annual_Income'].astype(float)
train1['Monthly_Inhand_Salary'] = train1['Monthly_Inhand_Salary'].astype(float)
train1['Num_Bank_Accounts'] = train1['Num_Bank_Accounts'].astype(int)
train1['Num_Credit_Card']=train1['Num_Credit_Card'].astype(int)
train1['Interest_Rate']=train1['Interest_Rate'].astype(int)
train1['Num_of_Loan'] = train1['Num_of_Loan'].astype(int)
train1['Delay_from_due_date']=train1['Delay_from_due_date'].astype(int)
train1['Num_of_Delayed_Payment'] = train1['Num_of_Delayed_Payment'].astype(float)
train1['Changed_Credit_Limit'] = train1['Changed_Credit_Limit'].astype(float)
# train1['Num_Credit_Inquiries'] = train1['Num_Credit_Inquiries'].astype(int)
train1['Outstanding_Debt'] = train1['Outstanding_Debt'].astype(float)
train1['Credit_Utilization_Ratio']=train1['Credit_Utilization_Ratio'].astype(float)
train1['Total_EMI_per_month']=train1['Total_EMI_per_month'].astype(float)
train1['Monthly_Balance'] = train1['Monthly_Balance'].astype(float)
train1['Amount_invested_monthly'] = train1['Amount_invested_monthly'].astype(float)
```

```python
import pandas as pd

# Assuming your DataFrame is named df
# Fill missing values in 'Num_of_Delayed_Payment' within each group separately
train1['Num_of_Delayed_Payment'] = train1.groupby('Customer_ID')['Num_of_Delayed_Payment'].transform(lambda x: x.fillna(x.mean()))
train1['Credit_History_Age'] = train1.groupby('Customer_ID')['Credit_History_Age'].transform(lambda x: x.fillna(x.mode()))
train1['Amount_invested_monthly'] = train1.groupby('Customer_ID')['Amount_invested_monthly'].transform(lambda x: x.fillna(x.mean()))
train1['Num_Credit_Inquiries'] = train1.groupby('Customer_ID')['Num_Credit_Inquiries'].transform(lambda x: x.fillna(x.mean()))
train1['Monthly_Inhand_Salary'].fillna(train1['Monthly_Inhand_Salary'].mean(), inplace=True)
# train1['Type_of_Loan'].fillna(train1['Type_of_Loan'].mode()[0], inplace=True)
```
