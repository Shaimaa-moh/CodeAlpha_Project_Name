```python
drop_columns=['ID',"Customer_ID","Name","SSN","Credit_History_Age",'Payment_Behaviour',"Type_of_Loan","Age","Month"]
train1=train_data.drop(drop_columns,axis=1)
```

```python

# Deleting (-) and (_) in Column
train1['Annual_Income'] = train1['Annual_Income'].str.replace('-','')
train1['Annual_Income'] = train1['Annual_Income'].str.replace('_','')

# Deleting (-) and (_) in Column
train1['Num_of_Loan'] = train1['Num_of_Loan'].str.replace('-','')
train1['Num_of_Loan'] = train1['Num_of_Loan'].str.replace('_','')

# Deleting (-) and (_) in Column
train1['Changed_Credit_Limit'] = train1['Changed_Credit_Limit'].str.replace('-','0')
train1['Changed_Credit_Limit'] = train1['Changed_Credit_Limit'].str.replace('_','0')

# Deleting (-) and (_) in Column
train1['Outstanding_Debt'] = train1['Outstanding_Debt'].str.replace('-','')
train1['Outstanding_Debt'] = train1['Outstanding_Debt'].str.replace('_','')

# Deleting (-) and (_) in Column
train1['Monthly_Balance'] = train1['Monthly_Balance'].str.replace('-','')
train1['Monthly_Balance'] = train1['Monthly_Balance'].str.replace('_','')
```

```python

train1['Annual_Income'] = train1['Annual_Income'].astype(float)
train1['Num_of_Loan'] = train1['Num_of_Loan'].astype(int)
train1['Changed_Credit_Limit'] = train1['Changed_Credit_Limit'].astype(float)
train1['Outstanding_Debt'] = train1['Outstanding_Debt'].astype(float)
train1['Monthly_Balance'] = train1['Monthly_Balance'].astype(float)
```

```python
train1['Monthly_Balance']=train1['Monthly_Balance'].fillna(train1['Monthly_Balance'].mean())
train1['Num_Credit_Inquiries']=train1['Num_Credit_Inquiries'].fillna(train1['Num_Credit_Inquiries'].mean())
```
